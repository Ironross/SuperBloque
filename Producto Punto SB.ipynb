```python
import pandas as pd

def extract_month_year(date):
    if '-' in date:  # Si ya está en formato "MM-AA"
        return date[:7]  # Extraer "MM-AA" directamente
    else:  # Si es una fecha completa
        return date[6:7] + '-' + date[9:10]  # Extraer "MM-AA" desde una fecha completa
```

```python
# Crear un DataFrame de ejemplo
data = {'Fecha': ['2024-10-21 00:00:00', '2023-11-15 00:00:00', '2025-09-05 00:00:00', 'Aug-21']}
df = pd.DataFrame(data)

# Aplicar la función a la columna "Fecha"
df['Mes_Año'] = df['Fecha'].apply(extract_month_year)

# Mostrar el DataFrame resultante
print(df)
```
